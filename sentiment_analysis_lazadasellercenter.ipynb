```python
# ======================================
# üìò Load kamus slang dari Excel
# ======================================
kamus = pd.read_excel("kamuskatabaku.xlsx")
slang_dict = dict(zip(kamus['tidak_baku'], kamus['kata_baku']))

def normalize_text(text):
    words = text.split()
    normalized_words = [slang_dict.get(word, word) for word in words]
    return " ".join(normalized_words)

def preprocessing_pipeline(text):
    # 1. Pembersihan awal (hapus angka, simbol, link)
    cleaned_text = re.sub(r'http\S+|www\S+|https\S+', '', text)
    cleaned_text = re.sub(r'[^a-zA-Z\s]', '', cleaned_text)

    # 2. Case Folding
    case_folding = cleaned_text.lower()

    # 3. Normalisasi (hasil tetap kalimat)
    normalized = normalize_text(case_folding)

    # 4. Tokenisasi
    tokens = nltk.word_tokenize(normalized)

    # 5. Stopword Removal
    stop_removed = [word for word in tokens if word not in stop_words]

    # 6. Stemming
    stemmed = [stemmer.stem(word) for word in stop_removed]

    return cleaned_text, case_folding, normalized, tokens, stop_removed, stemmed, " ".join(stemmed)

# ======================================
# üìä PREPROCESSING
# ======================================
ls_df[['cleaned_text','case_folding','normalized','tokenized','stop_removed','stemmed','final_cleaned']] = ls_df['content'].astype(str).apply(
    lambda x: pd.Series(preprocessing_pipeline(x))
)

# ======================================
# üîç Lihat hasil
# ======================================
print(ls_df[['content','cleaned_text','case_folding','normalized','tokenized','stop_removed','stemmed','final_cleaned']].head(11))
```

```python
# =====================================
# üìö IMPORT LIBRARY
# =====================================
import matplotlib.pyplot as plt

# =====================================
# üìÅ BACA KAMUS POSITIF & NEGATIF DENGAN SKOR (PASTIKAN TIPE DATA ANGKA)
# =====================================
positive_words = pd.read_csv('/content/positive.tsv', sep='\t', header=None, names=['word', 'score'])
negative_words = pd.read_csv('/content/negative.tsv', sep='\t', header=None, names=['word', 'score'])

# Konversi kolom skor ke numerik
positive_words['score'] = pd.to_numeric(positive_words['score'], errors='coerce').fillna(0)
negative_words['score'] = pd.to_numeric(negative_words['score'], errors='coerce').fillna(0)

# Gabungkan kamus
lexicon_df = pd.concat([positive_words, negative_words], ignore_index=True)
lexicon_dict = dict(zip(lexicon_df['word'].astype(str).str.lower(), lexicon_df['score']))

print(f"Total kata dalam lexicon gabungan: {len(lexicon_dict)}")

# =====================================
# üß© PASTIKAN KOLOM STEMMED BERUPA LIST TOKEN
# =====================================
if ls_df['stemmed'].apply(lambda x: isinstance(x, str)).any():
    ls_df['stemmed'] = ls_df['stemmed'].apply(lambda x: x.split())

# =====================================
# üß† DEFINISI FUNGSI LABELING BERBOBOT
# =====================================
def label_sentimen_berbobot(stemmed_tokens):
    if not isinstance(stemmed_tokens, list):
        return "netral"

    total_skor = 0.0
    for kata in stemmed_tokens:
        kata = kata.lower()
        skor = lexicon_dict.get(kata, 0)
        try:
            total_skor += float(skor)
        except:
            continue

    if total_skor > 0:
        return "positif"
    elif total_skor < 0:
        return "negatif"
    else:
        return "netral"

# =====================================
# üè∑Ô∏è PENERAPAN LABELING
# =====================================
ls_df['sentimen'] = ls_df['stemmed'].apply(label_sentimen_berbobot)

# =====================================
# ‚ùå HAPUS DATA NETRAL
# =====================================
before_len = len(ls_df)
ls_df = ls_df[ls_df['sentimen'] != 'netral'].reset_index(drop=True)
after_len = len(ls_df)

print(f"\nüßπ Menghapus data netral...")
print(f"Jumlah data sebelum dihapus: {before_len}")
print(f"Jumlah data sesudah dihapus: {after_len}")
print(f"Total data netral yang dihapus: {before_len - after_len}\n")

# =====================================
# üìä DISTRIBUSI SENTIMEN (SETELAH HAPUS NETRAL)
# =====================================
sentiment_counts = ls_df['sentimen'].value_counts()

print("=== üìà DISTRIBUSI SENTIMEN (SETELAH HAPUS NETRAL) ===")
print(sentiment_counts)

# =====================================
# üé® GRAFIK DISTRIBUSI SENTIMEN
# =====================================
warna = {'positif': 'green', 'negatif': 'red'}

plt.figure(figsize=(6,4))
plt.bar(sentiment_counts.index,
        sentiment_counts.values,
        color=[warna.get(label, 'gray') for label in sentiment_counts.index])
plt.title('Distribusi Sentimen Positif & Negatif Berdasarkan Kamus Lexicon', fontsize=12, fontweight='bold')
plt.xlabel('Kategori Sentimen')
plt.ylabel('Jumlah Data')
plt.grid(axis='y', linestyle='--', alpha=0.6)

for i, v in enumerate(sentiment_counts.values):
    plt.text(i, v + 0.5, str(v), ha='center', fontweight='bold')

plt.show()

# =====================================
# üìã CONTOH DATA PER KATEGORI
# =====================================
print("\n=== üíñ Contoh Data POSITIF ===")
print(ls_df[ls_df['sentimen']=='positif'][['content','cleaned_text','case_folding','normalized','tokenized','stop_removed','stemmed','final_cleaned','sentimen']].head(5))

print("\n=== üí¢ Contoh Data NEGATIF ===")
print(ls_df[ls_df['sentimen']=='negatif'][['content','cleaned_text','case_folding','normalized','tokenized','stop_removed','stemmed','final_cleaned','sentimen']].head(5))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from imblearn.over_sampling import ADASYN
from sklearn.svm import SVC
import seaborn as sns
import matplotlib.pyplot as plt

# -------------------------
# üîß Fungsi pipeline tuning SVM dengan ADASYN + Confusion Matrix terbaik
# -------------------------
def run_pipeline(test_size, C_values=[0.1, 1, 10]):
    print("\n==============================")
    print(f"üîÄ Skenario Split Dataset ‚Üí test_size={test_size}")
    print("==============================")

    # ‚úÖ Ganti df jadi ls_df_nonetral
    X_train, X_test, y_train, y_test = train_test_split(
        ls_df['final_cleaned'], ls_df['sentimen'],
        test_size=test_size, random_state=42, stratify=ls_df['sentimen']
    )

    print(f"üì¶ Training samples awal: {X_train.shape[0]}")
    print(f"üì¶ Testing samples: {X_test.shape[0]}")

    # TF-IDF (fit hanya di training, transform di test)
    tfidf = TfidfVectorizer(max_features=5000)
    X_train_tfidf = tfidf.fit_transform(X_train)
    X_test_tfidf = tfidf.transform(X_test)

    # ADASYN hanya di data training ‚Üí tidak ada leakage
    adasyn = ADASYN(random_state=42)
    X_train_res, y_train_res = adasyn.fit_resample(X_train_tfidf, y_train)
    print(f"üì¶ Training samples setelah ADASYN: {X_train_res.shape[0]}")

    # -------------------
    # üîç Tuning parameter C
    # -------------------
    best_acc = 0
    best_c = None
    best_report = None
    best_y_pred = None

    for c in C_values:
        model = SVC(kernel='linear', C=c, random_state=42)
        model.fit(X_train_res, y_train_res)
        y_pred = model.predict(X_test_tfidf)

        acc = accuracy_score(y_test, y_pred)
        print(f"\nüß© C={c} ‚Üí Accuracy: {acc:.4f}")

        if acc > best_acc:
            best_acc = acc
            best_c = c
            best_report = classification_report(y_test, y_pred, digits=4)
            best_y_pred = y_pred

    # -------------------
    # üèÜ Hasil terbaik
    # -------------------
    print("\n‚úÖ Hasil Terbaik untuk Split", test_size)
    print(f"üåü Nilai C terbaik: {best_c}")
    print(f"üéØ Akurasi tertinggi: {best_acc:.4f}")
    print("\nüìä Classification Report (C terbaik):")
    print(best_report)

    # -------------------
    # üìâ Confusion Matrix dari C terbaik
    # -------------------
    labels = sorted(ls_df['sentimen'].unique())
    cm = confusion_matrix(y_test, best_y_pred, labels=labels)

    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Greens',
                xticklabels=labels, yticklabels=labels)
    plt.title(f"Confusion Matrix (C={best_c}, test_size={test_size})", fontsize=12, weight='bold')
    plt.xlabel("Predicted Label")
    plt.ylabel("Actual Label")
    plt.tight_layout()
    plt.show()

# =========================
# üîπ Jalankan beberapa skenario split
# =========================
run_pipeline(0.1)
run_pipeline(0.2)
run_pipeline(0.3)
run_pipeline(0.4)
```

```python
from wordcloud import WordCloud

def plot_tfidf_wordcloud(df, label_value, label_name):
    subset = df[df['sentimen'] == label_value]
    texts = subset['stemmed'].astype(str).tolist()

    if len(texts) == 0:
        print(f"Tidak ada data untuk label '{label_name}'")
        return

    vectorizer = TfidfVectorizer(
        max_features=5000,
        token_pattern=r"(?u)\b\w+\b"  # pastikan kata tunggal terbaca
    )
    X_tfidf = vectorizer.fit_transform(texts)
    feature_names = vectorizer.get_feature_names_out()
    tfidf_scores = np.asarray(X_tfidf.mean(axis=0)).ravel()
    tfidf_dict = dict(zip(feature_names, tfidf_scores))

    wordcloud = WordCloud(
        width=1000,
        height=500,
        background_color='white',
        colormap='viridis'
    ).generate_from_frequencies(tfidf_dict)

    plt.figure(figsize=(12, 6))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(f"WordCloud Berdasarkan Sentimen {label_name}", fontsize=16, weight="bold")
    plt.show()

plot_tfidf_wordcloud(ls_df, "positif", "Positif")
plot_tfidf_wordcloud(ls_df, "negatif", "Negatif")
```
